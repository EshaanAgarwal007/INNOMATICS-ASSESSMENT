
Problem Statement 1: Employee Performance Bonus Eligibility

Input:
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

Expected Output:
Top Performers Eligible for Bonus: Ravi, Kiran (Score: 92)





```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
highest=0
for i in employees:
  if employees[i]>highest:
    highest=employees[i]
print("Top Performers Eligible for Bonus:", end=" ")
for i in employees:
    if employees[i] == highest:
        print(i, end=", ")

print("(Score:", highest ,")")
```

Problem Statement 2: Search Query Keyword Analysis

Input:
"Buy mobile phone buy phone online"

Expected Output:
{'buy': 2, 'phone': 2}



```python
n =str(input())
n = n.lower()
words = n.split()
freq = {}
for w in words:
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1
result = {}
for key in freq:
    if freq[key] > 1:
        result[key] = freq[key]
print(result)
```

Problem Statement 3: Sensor Data Validation

Input:
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

Expected Output:
Valid Sensor Readings (Hour, Value):
[(1, 4), (3, 8), (4, 10), (5, 12)]



```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

Problem Statement 4: Email Domain Usage Analysis

Input:
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]

Expected Output:
gmail.com: 60%
yahoo.com: 40%



```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}
for i in emails:
    domain = ""
    found = False
    for ch in i:
        if ch == "@":
            found = True
            continue
        if found == True:
            domain = domain + ch
    if domain in domain_count:
        domain_count[domain] = domain_count[domain] + 1
    else:
        domain_count[domain] = 1
total = len(emails)
print("Domain Usage:")
for d in domain_count:
    percent = (domain_count[d] * 100) // total
    print(d + ":", percent, "%")
```

Problem Statement 5: Sales Spike Detection

Input:
sales = [1200, 1500, 900, 2200, 1400, 3000]

Expected Output:
Day 4: 2200
Day 6: 3000


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
total = 0
for i in sales:
    total = total + i
avg = total / len(sales)
for i in range(len(sales)):
    if sales[i] > avg:
        print("Day", i+1, ":", sales[i])
```

Problem Statement 6: Duplicate User ID Detection

Input:
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

Expected Output:
user1 → 3 times
user3 → 2 times


```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
count_dict = {}
for uid in user_ids:
    if uid in count_dict:
        count_dict[uid] = count_dict[uid] + 1
    else:
        count_dict[uid] = 1
for key in count_dict:
    if count_dict[key] > 1:
        print(key, "→", count_dict[key], "times")
```
