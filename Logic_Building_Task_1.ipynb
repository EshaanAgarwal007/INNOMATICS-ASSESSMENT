PASS/FAIL ANALYZER

```python
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for i in marks:
  if i>=50:
    print("Student Pass")
    pass_count=pass_count+1
  else:
    print("Student Fail")
    fail_count=fail_count+1
print("Total Pass Count:", pass_count)
print("Total Fail Count:", fail_count)
```

SIMPLE DATA CLEANER

```python
names = [" Alice ", "bob", " CHARLIE "]
print(f"names before changing:{names}")
for i in range(len(names)):
  names[i]=names[i].strip()
  names[i]=names[i].lower()
print(f"names after changing to lower case:{names}")
```

MESSAGE LENGTH ANALYZER

```python
n=int(input())
message=[]
for i in range(n):
  val=input()
  message.append(val)
for i in message:
  if(len(i)>10):
    print(f"{i}-->Message Accepted,length of message {len(i)}")
  else:
    print(f"{i}-->Message not Accepted,length of message {len(i)}")
```

ERROR MESSAGE DETECTOR

```python
n=int(input())
logs=[]
count=0
for i in range(n):
  val=input()
  logs.append(val)
for i in logs:
  if(i=="ERROR"):
    count+=1
print(f"Number of ERROR entries are: {count}")
```
